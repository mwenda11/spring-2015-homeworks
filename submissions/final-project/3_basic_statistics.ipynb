```python
# to predict default -- filter: paid,defaulted
dframe = dframe[dframe.status != "expired"]
dframe = dframe[dframe.status != "in_repayment"]
dframe = dframe[dframe.status != "refunded"]
```

```python
# to binary
def is_paid(row):
    paid = row.status
    if paid == "paid":
        return 1
    else:
        return 0

dframe['status'] = dframe.apply(is_paid, axis = 1)


for it, dat in dframe.groupby("status"):
    print it, len(dat)
```

```python
# to binary
def is_female(row):
    gen = row.borrower_gender
    if gen == "F":
        return 1
    else:
        return 0

dframe['borrower_gender'] = dframe.apply(is_female, axis = 1)

for it, dat in dframe.groupby("borrower_gender"):
    print it, len(dat)
```

```python
# add continent
asia = ["Afghanistan","Armenia","Azerbaijan","Cambodia","Georgia","Indonesia","Iraq","Israel","Jordan","Kyrgyzstan","Lebanon","Moldova","Mongolia","Nepal","Pakistan","Palestine","Philippines","Samoa","Sri Lanka","Tajikistan","Turkey","Vietnam","Yemen"]
africa = ["Benin","Burkina Faso","Burundi","Cameroon","Congo (Dem. Rep.)","Congo (Rep.)","Cote D'Ivoire","Ghana","Kenya","Liberia","Mali","Mozambique","Nigeria","Rwanda","Senegal","Sierra Leone","South Africa","South Sudan","Tanzania","Togo","Uganda","Zimbabwe"]
latin_america = ["Bolivia","Colombia","Chile","Costa Rica","Dominican Republic","Ecuador","El Salvador","Guatemala","Haiti","Honduras","Mexico","Nicaragua","Paraguay","Peru"]
europe = ["Bosnia and Herzegovina","Bulgaria","Ukraine"]
north_america = ["United States"]
```

```python
def continent(row):
    con = row.country.strip()
    if con in africa:
        return 'africa'
    elif con in asia:
        return 'asia'
    elif con in europe:
        return 'europe'
    elif con in latin_america:
        return 'latin_america'
    elif con in north_america:
        return 'north_america'

dframe.insert(8, "continent", '')
dframe['continent'] = dframe.apply(continent, axis = 1)

for it, dat in dframe.groupby("continent"):
    print it, len(dat)
```

**Loan flow: lender countries to recipients** <br>
For each major source country, identify recipient receiving most loans. <br>
Checking if the lenders from particular countries have prefences on which countries to lend to.

```python
xframe = pd.read_csv('loan_src2dest.csv')
```

```python
xframe.num_loans.max()
```

```python
loan_matrix = np.zeros((201,4), dtype='S6')
idx = 0
for source, src_data in xframe.groupby("source"):
    max_dest = '' # recipient with most loans
    max_dest_count = 0 # num_loans to max_dest
    total_src_loans = 0 # total loans from source country

    for indx, row in src_data.iterrows():      
        count = row.num_loans
        total_src_loans = total_src_loans + count
        if count > max_dest_count:
            max_dest_count = count
            max_dest = row.destination
    #print source, max_dest, max_dest_count, total_src_loans
    loan_matrix[idx][0] = source
    loan_matrix[idx][1] = max_dest
    loan_matrix[idx][2] = max_dest_count
    loan_matrix[idx][3] = total_src_loans
    
    idx += 1

idx
```

```python
kframe = pd.DataFrame(loan_matrix, columns=["source","highest_recipient","highest_recipient_loans","total_from_source"])
kframe.highest_recipient_loans = kframe['highest_recipient_loans'].astype(int)
kframe.total_from_source = kframe['total_from_source'].astype(int)
```
