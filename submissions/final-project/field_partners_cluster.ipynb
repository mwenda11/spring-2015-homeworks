```python
pframe = pd.read_csv('partners.csv', index_col='pid', parse_dates=['start_date'])
```

```python
def active_days(row):
    start = row.start_date
    now = datetime.datetime.now()
    diff = (now - start)
    return diff.days

pframe['days_active'] = pframe.apply(active_days, axis=1)
```
