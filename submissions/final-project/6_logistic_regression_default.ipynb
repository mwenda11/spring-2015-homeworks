**Predicting if a loan gets paid**

```python
import os
import json
import time
import datetime
import requests
import numpy as np
import pandas as pd

from patsy import dmatrices
from collections import defaultdict
from matplotlib import pyplot as plt
from pandas import Series, DataFrame
%matplotlib inline

import sklearn.metrics as metrics
import sklearn.utils as utils
import sklearn.linear_model as linear_model
import sklearn.cross_validation as cross_validation
from sklearn.preprocessing import StandardScaler
```

```python
dframe = pd.read_csv('filtered_loan_dataD.csv', index_col='id')
```

```python
#dframe.head()
```

```python
# to predict default -- filter: paid,defaulted
dframe = dframe[dframe.status != "expired"]
dframe = dframe[dframe.status != "in_repayment"]
dframe = dframe[dframe.status != "refunded"]
```

```python
# convert to binary paid or not
def is_paid(row):
    paid = row.status
    if paid == "paid":
        return 1
    else:
        return 0

dframe['status'] = dframe.apply(is_paid, axis = 1)


for it, dat in dframe.groupby("status"):
    print it, len(dat)
```

```python
# convert to binary. 0=Male, 1=Female
def is_female(row):
    gen = row.borrower_gender
    if gen == "F":
        return 1
    else:
        return 0

dframe['borrower_gender'] = dframe.apply(is_female, axis = 1)

for it, dat in dframe.groupby("borrower_gender"):
    print it, len(dat)
```

```python
# add column continent
con_list = []
for it, dat in dframe.groupby("country"):
    con_list.append(it.strip())

#print con_list
asia = ["Afghanistan","Armenia","Azerbaijan","Cambodia","Georgia","Indonesia","Iraq","Israel","Jordan","Kyrgyzstan","Lebanon","Moldova","Mongolia","Nepal","Pakistan","Palestine","Philippines","Samoa","Sri Lanka","Tajikistan","Turkey","Vietnam","Yemen"]
africa = ["Benin","Burkina Faso","Burundi","Cameroon","Congo (Dem. Rep.)","Congo (Rep.)","Cote D'Ivoire","Ghana","Kenya","Liberia","Mali","Mozambique","Nigeria","Rwanda","Senegal","Sierra Leone","South Africa","South Sudan","Tanzania","Togo","Uganda","Zimbabwe"]
latin_america = ["Bolivia","Colombia","Chile","Costa Rica","Dominican Republic","Ecuador","El Salvador","Guatemala","Haiti","Honduras","Mexico","Nicaragua","Paraguay","Peru"]
europe = ["Bosnia and Herzegovina","Bulgaria","Ukraine"]
north_america = ["United States"]
```

```python
def continent(row):
    con = row.country.strip()
    if con in africa:
        return 'africa'
    elif con in asia:
        return 'asia'
    elif con in europe:
        return 'europe'
    elif con in latin_america:
        return 'latin_america'
    elif con in north_america:
        return 'north_america'

dframe.insert(8, "continent", '')
dframe['continent'] = dframe.apply(continent, axis = 1)

for it, dat in dframe.groupby("continent"):
    print it, len(dat)
```

**Logistic regression model to predict if a loan is paid or not (defaulted)**

```python
y, dfX = dmatrices('status ~ text_length + C(sector) + C(continent) + loan_amount + lender_count + bonus_credit_elig + num_borrowers + \
                 borrower_gender + partner_loss_liable + days_fundraising + translator + pactive + prating + pdays_as_member + pdelinquency_rate + \
                 pdefault_rate + ptotal_amount_raised + ploans_posted + pcharges_FI + pALSPCI + ploans_at_risk_rate + pforex_loss_rate',
                 dframe, return_type="dataframe")

# high correlation between 'paid' and the columns below so exclude them
# delinquent
# days_late
```

```python
# center data
scaler = StandardScaler()
scaler.fit(dfX)
X = scaler.transform(dfX)
```

```python
dframe.iloc[123:128,:10]
```

```python
print y.shape, 
y = np.ravel(y)
print y.shape
```

```python
# evaluate the model by splitting into train and test sets
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X,
                                                                     y,
                                                                     test_size=0.3,
                                                                     random_state=0
                                                                     )
logistic_regr = linear_model.LogisticRegression()
logistic_regr.fit(X_train, y_train)
```

```python
# verifying if cross_validation did a good split
dfy_train = pd.DataFrame(y_train, columns=["status"])
dfy_test = pd.DataFrame(y_test, columns=["status"])

for it, dat in dfy_test.groupby("status"):
    print it, len(dat)
print '----------------------'   
for it, dat in dfy_train.groupby("status"):
    print it, len(dat)
```

```python
# predict class labels for the test set
y_predicted = logistic_regr.predict(X_test)
print y_predicted
```

```python
#prediction probabilities
probs = logistic_regr.predict_proba(X_test)
print probs
```

```python
# generate evaluation metrics
print metrics.accuracy_score(y_test, y_predicted)
```

```python
print metrics.confusion_matrix(y_test, y_predicted)
print metrics.classification_report(y_test, y_predicted)
```

```python
# examine the coefficients
print pd.DataFrame(zip(dfX.columns, np.transpose(logistic_regr.coef_)))
```

**Summary**

- As expected, there is a high correlation between the defaulting of a loan and the partner's default rate.
- When the field partner who posted a loan is liable for loss, meaning that in the event of a default it's the partner and not the lender who will incur a loss, the likelihood of default is low. This can be because the partner has incentive pursue the borrower if the borrower defaults.
- Loans to Asia were the least likely to be defaulted on compared to other continents.
- Higher loan amounts are more likely to be defaulted on.
