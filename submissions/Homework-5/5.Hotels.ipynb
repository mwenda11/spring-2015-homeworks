** Task 2 (30 pts)**

Then, for each hotel that our search returns, we will "click" (with the code of course) on it and scrape the information below.

![Information to be scraped](hotel_info.png)

Of course, feel free to collect even more data if you want. 

````python
# return a list of hrefs (href for each hotel)
url_list = scrape.get_url_list("boston","ma")
````

````python
# For each link, 'click it' and parse the data into a 2d array.
# Takes about 4 minutes 45 seconds.

print len(url_list)
base_url = "http://www.tripadvisor.com"
user_agent = "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2272.76 Safari/537.36"
datadir = "data/"

columns = ["overall_stars","r_excellent","r_very_good","r_average","r_poor","r_terrible","family_reviews","couples_reviews","solo_reviews","business_reviews","location_rating","sleep_quality_rating","rooms_rating","service_rating","value_rating","cleanliness_rating"]

data_matrix = np.zeros((len(url_list),len(columns)), dtype=np.float)

for row_idx, hotel_url in enumerate(url_list):
    url = base_url + hotel_url
    # Sleep 2 sec before starting a new http request
    time.sleep(2)
    # Request page
    headers = { 'User-Agent' : user_agent }
    response = requests.get(url, headers=headers)
    html = response.text.encode('utf-8')
    # Save the webpage ?
    
    soup = BeautifulSoup(html)
    
    # Get overall rating
    top_box = soup.find('div', {'id': 'HEADING_GROUP'})
    rating_span = name = top_box.find('span', {'class': 'rate sprite-rating_rr rating_rr'})
    
    overall_stars_ele = rating_span.find('img')
    overall_stars = str(overall_stars_ele['alt'].split()[0])
    data_matrix[row_idx,0] = float(overall_stars.strip())
    
    # Get rating break-down: Excellent, Very Good, Average,...
    ratings_box = soup.find('div', {'class' :'content wrap trip_type_layout'})
    rating_eles = ratings_box.findAll("span", {'class': 'compositeCount'})
    
    col_idx = 1            
    for rating_ele in rating_eles:         
        val = rating_ele.find(text=True)
        val = str(val)
        val = val.replace(",","")
        val = float(val)        
        data_matrix[row_idx,col_idx] = val
        col_idx +=1
    
    # Rater groupings: Families, Couples, ...
    review_eles = ratings_box.findAll('div', {'class': 'filter_connection_wrapper'})
    
    for review_ele in review_eles:         
        val = review_ele.find('div', {'class': 'value'}).find(text=True)
        val = str(val)
        val = val.replace(",","")
        val = float(val)
        data_matrix[row_idx,col_idx] = val
        col_idx +=1
    
    # Get rating summary for various categories: sleep quality, location, room, ...
    summary_box = ratings_box.find('div', {'id': 'SUMMARYBOX'})
    summary_list = summary_box.findAll('li')
    
    for category in summary_list:
        name = category.find('div', {'class': 'name'}).find(text=True)
        stars_elem = category.find('img')
        stars = stars_elem['alt'].split()[0]
        
        name = str(name)
        name = name.strip().lower()
        
        if name == "location":
            col_idx = columns.index("location_rating")
        elif name == "sleep quality":
            col_idx = columns.index("sleep_quality_rating")
        elif name == "rooms":
            col_idx = columns.index("rooms_rating")
        elif name == "service":
            col_idx = columns.index("service_rating")
        elif name == "value":
            col_idx = columns.index("value_rating")
        elif name == "cleanliness":
            col_idx = columns.index("cleanliness_rating")
            
        val = float(stars.strip())
        data_matrix[row_idx,col_idx] = val

print "DONE"
````

````python
# create a dataframe from the data matrix
dframe = pd.DataFrame(data_matrix, columns = columns)
dframe.iloc[0:10,:]
````

** Task 4 (30 pts) **

Finally, we will use logistic regression to decide if a hotel is _excellent_ or not. We classify a hotel as _excellent_ if more than **60%** of its ratings are 5 stars. This is a binary attribute on which we can fit a logistic regression model. As before, use the model to analyze the data.

-------

````python
def is_excellent(row):  
    val = row.r_excellent/row.total_reviews
    if val > 0.60:
        return 1
    else:
        return 0
````

````python
dframe = pd.DataFrame(data_matrix, columns = columns)

dframe.insert(1, 'total_reviews', 0.0) # add new column
dframe['total_reviews'] = dframe.r_excellent + dframe.r_very_good + dframe.r_average + dframe.r_poor + dframe.r_terrible

dframe.insert(1, 'excellent?', 0.0) # add new column
dframe['excellent?'] = dframe.apply(is_excellent, axis = 1)

dframe.iloc[0:10,:]
````

````python
dframe.groupby('excellent?').mean()
````

````python
train_cols = dframe.columns[8:] #Perfect separation occured if we include the ratings categories: excellent, very good
 
#logit = sm.Logit(dframe['excellent?'], dframe[train_cols])
# fit the model
#result = logit.fit() 

X = dframe[train_cols]
y_frame = dframe['excellent?']
y = np.ravel(y_frame)
````

````python
# evaluate the model by splitting into train and test sets
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X,
                                                                     y,
                                                                     test_size=0.3,
                                                                     random_state=0
                                                                     )
logistic_regr = linear_model.LogisticRegression()
logistic_regr.fit(X_train, y_train)
````

````python
# predict class labels for the test set
y_predicted = logistic_regr.predict(X_test)
print y_predicted
````

````python
#prediction probabilities
probs = logistic_regr.predict_proba(X_test)
print probs
````

````python
# generate evaluation metrics
print metrics.accuracy_score(y_test, y_predicted)
````

````python
print metrics.confusion_matrix(y_test, y_predicted)
print metrics.classification_report(y_test, y_predicted)
````
