```python
import numpy as np
import pandas as pd
import scipy as sp
import scipy.cluster.hierarchy as hr
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.spatial.distance import pdist, squareform
%matplotlib inline
```

```python
countries = set()
with open("allr_recipes.txt", "r") as f:
    for line in f:
        tokens = line.split()
        countries.add(tokens[0])

print len(countries) 
print sorted(countries)
```

```python
regions = set()
with open("epic_recipes.txt", "r") as f:
    for line in f:
        tokens = line.split()
        regions.add(tokens[0])

print len(regions) 
print sorted(regions)
```

```python
cuisine = set()
with open("menu_recipes.txt", "r") as f:
    for line in f:
        tokens = line.split()
        cuisine.add(tokens[0])

print len(cuisine) 
print sorted(cuisine)
```

```python
# parse recipe files to get list of recipes and set of ingredients
ingredients = set()
recipes = []

with open("allr_recipes.txt", "r") as f:
    for line in f:
        recipes.append(line)
        tokens = line.split()
        for t in tokens[1:]:
            ingredients.add(t)
            
with open("epic_recipes.txt", "r") as f:
    for line in f:
        recipes.append(line)
        tokens = line.split()
        for t in tokens[1:]:
            ingredients.add(t)
            
with open("menu_recipes.txt", "r") as f:
    for line in f:
        recipes.append(line)
        tokens = line.split()
        for t in tokens[1:]:
            ingredients.add(t)            

print len(ingredients)
print len(recipes)
```

```python
country_dict = {}
#read dict file and put to map
with open("countries_dict.txt", "r") as f:
    for line in f:
        tokens = line.split()
        country_dict[tokens[0]] = tokens[1]

print country_dict
```

```python
columns = tuple(ingredients)
columns = sorted(columns)
# print ingredients
country_region_list = []
    
binary_matrix = np.zeros((len(recipes),len(columns)), dtype=np.int) 

row_idx = 0
for recipe in recipes:
    tokens = recipe.split()
    country_region = country_dict[tokens[0].lower()]
    country_region_list.append(country_region)
    for t in tokens[1:]:
        col_idx = columns.index(t)
        binary_matrix[row_idx][col_idx] = 1
    row_idx = row_idx + 1
    
print "done creating matrix..."

dframe = pd.DataFrame(binary_matrix, columns = columns)
dframe.insert(0, 'country_region', 'country_region') # add new column

dframe.country_region = country_region_list

dframe.iloc[0:10,0:10]
```

```python
#dframe.iloc[len(recipes)-20:,20:40]
#dframe.iloc[0:10,0:20]
#binary_matrix[0:10,0:20]
```

```python
# Group recipes by country_region
country_set = set()
for country, country_data in dframe.groupby("country_region"):
    country_set.add(country)
    
print len(country_set)
print len(columns)

grouped_matrix = np.zeros((len(country_set),len(columns)),dtype=np.float)
country_reg_list = []

r_idx = 0
for country, country_data in dframe.groupby("country_region"):
    country_reg_list.append(country)
    num_recipes = len(country_data)*1.0
    #print num_recipes
    for col in columns:
        c_idx = columns.index(col)
        val = (country_data[col].sum())/num_recipes #normalize
        grouped_matrix[r_idx][c_idx] = val
    r_idx = r_idx + 1
    
#grouped_matrix[0:10,0:15]
grouped_frame = pd.DataFrame(grouped_matrix, columns = columns)
grouped_frame.insert(0, 'country_region', 'country_region') # add new column

grouped_frame.country_region = country_reg_list

grouped_frame.iloc[0:10,0:10]
```

```python
test = grouped_frame.iloc[0:len(country_reg_list),1:len(columns)]
test.iloc[0:10,0:10]
```

```python
Z = hr.linkage(test, method='complete', metric='euclidean')

print Z.shape, test.shape
print country_reg_list
```

```python
fig = plt.figure(figsize=(20,10))
T = hr.dendrogram(Z,color_threshold=0.7, leaf_font_size=12,labels=country_reg_list)
fig.show()
```

I found it interesting that India, Malaysia, and South-Africa, which are in different continents, are in the same cluster. So I decided to analyze this particular cluster. One reason for the similarity in cuisine could be that malaysian and south-african cuisines are influenced by indians, who are a minority in both countries. 6.7 percent of population in malaysia and 2.5 percent of population in South-Africa is indian (according to Wikipendia).
To look more closely in this similarity, I retrieve the top ten most popular ingredient in each of the three countries, and plot a graph for each.

```python
india_malay_sa = grouped_frame[(grouped_frame.country_region == 'malaysia') | (grouped_frame.country_region == 'india') | (grouped_frame.country_region == 'south-africa')]
row_india = india_malay_sa.iloc[0,1:]
row_india_sort = np.argsort(row_india)

india10 = []
for d in range(373,383):
    india10.append(grouped_frame.columns[row_india_sort[d]+1])

print india10

row_malay = india_malay_sa.iloc[1,1:]
row_malay_sort = np.argsort(row_malay)

malay10 = []
for d in range(373,383):
    malay10.append(grouped_frame.columns[row_malay_sort[d]+1])

print malay10

row_sa = india_malay_sa.iloc[2,1:]
row_sa_sort = np.argsort(row_sa)

sa10 = []
for d in range(373,383):
    sa10.append(grouped_frame.columns[row_sa_sort[d]+1])

print sa10

ims
```

```python
top_ings = set(india10 + sa10 + malay10)
top_ings = list(top_ings)

select_countries = india_malay_sa[top_ings]

# 
for country in select_countries.index:
    select_countries.loc[country].plot(label = country, figsize =(15,5),rot=45)
plt.legend(loc='best')

#print (top_ings)
select_cols = ['country_region'] + top_ings
select_countries = india_malay_sa[select_cols]
select_countries
```

From the above we see a close similarity in ingredients, especially for India and Malaysia which are in the same sub-cluster.
